```python

#coder

def coder(arr):
    print("input array",arr)
    dup=[]

    for i in range(len(arr)):
        for j in range(i+1,len(arr),1):
            if (arr[i] ==arr[j] and arr[i] not in dup ):
                dup.append(arr[i])
            
    print(dup)
    
    return None
arr=list(map(int,input().split()))
res=coder(arr)




```

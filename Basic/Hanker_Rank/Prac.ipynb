```python
# Input binary string
binary_string = "000001001011101010010110011"

# Function to convert binary to DNA sequence
def binary_to_dna(binary_str):
    # Define the mapping of codons to amino acids
    codon_mapping = {
        '000': 'A', '001': 'C', '010': 'G', '011': 'T',
        '100': 'A', '101': 'C', '110': 'G', '111': 'T'
    }

    # Split the binary string into groups of three
    codons = [binary_str[i:i+3] for i in range(0, len(binary_str), 3)]

    # Convert each codon to its corresponding amino acid
    dna_sequence = ''.join(codon_mapping[codon] for codon in codons)

    return dna_sequence

# Convert binary string to DNA sequence
dna_sequence = binary_to_dna(binary_string)

# Output the result
print("Input Binary String:", binary_string)
print("Converted DNA Sequence:", dna_sequence)
```

```python




import mysql.connector

con=mysql.connector.connect(host="localhost",user="root",passwd="root",database="gauri")

c=con.cursor()

a="insert into student(rollno,name,city)values(%s,%s,%s)"
b=[(1,"saikrushna","mumbai"),(2,"dhanu","worli"),(3,"manoj","sion")]

try:
    c.executemany(a,b)
    con.commit()
except:
    con.rollback()


print(c.rowcount,"affected")

con.close()
```
